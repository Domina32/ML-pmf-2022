## [NumPy](https://numpy.org/)
>"The fundamental package for scientific computing with Python"

```python
import numpy as np

x = np.array([[1, 2, 3], [4, 5, 6]])
print("x:\n{}\n".format(x))
print("type:", type(x))
print("shape:", x.shape)
```

```python
# Return evenly spaced numbers over a specified interval.
x = np.linspace(-10, 10, 20)
print("x:\n{}\n".format(x))
print("type:", type(x))
print("shape:", x.shape)
```

## [Pandas](https://pandas.pydata.org/)
>"pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language."

```python
import pandas as pd
print("pandas version:", pd.__version__)

# create a simple dataset of people
data = {'Name': ["John", "Anna", "Peter", "Linda", "Mary"],
        'Location' : ["New York", "Paris", "Berlin", "London", "Paris"],
        'Age' : [24, 13, 53, 33, 18]
       }

data_pandas = pd.DataFrame(data)
# IPython.display allows "pretty printing" of dataframes
# in the Jupyter notebook
display(data_pandas)
```

```python
# Select all rows that have an age column greater than 30
display(data_pandas[data_pandas.Age > 30])
```

```python
# Select all rows that have a location column equal Paris
display(data_pandas[data_pandas["Location"] == "Paris"])
```

## [Matplotlib](https://matplotlib.org/)
> "Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible."

```python
import matplotlib
print("matplotlib version:", matplotlib.__version__)
```

```python
import matplotlib.pyplot as plt

# Generate a sequence of numbers from -10 to 10 with 100 steps in between
x = np.linspace(-10, 10, 100)

# Create a second array using sine
y = np.sin(x)

# The plot function makes a line chart of one array against another
plt.plot(x, y, marker="x")
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
  
# creating the dataset
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())
  
fig, ax = plt.subplots(1,1,figsize=(7,7))

# creating the bar plot
p1 = ax.bar(courses, values, color =['blue', 'red', 'green', 'orange'], label=courses,
        width = 0.7)
 
ax.set_xlabel("Courses offered")
ax.set_ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
ax.legend(p1, courses, loc=0)
plt.show()
```
